```python
# import libraries

import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# initialize random seed

np.random.seed(10)
```

```python
# generate random data

X = np.sort(np.random.uniform(low = 0, high = 1, size = 500))
Y = np.sin(2*np.pi*X)
noise = np.random.normal(scale = 0.3, size = Y.shape)
YY = Y + noise
```

```python
random.seed(10)
```

```python
# split data into training (randomly)

ind = random.sample(range(500), 500)
ind_train = ind[:50]
XX_train = X[ind_train]
YY_train = YY[ind_train]
```

```python
# make an object of Linear Regression and fit the train data
l1 = LinearRegression()
l1.fit(XX_train[:,np.newaxis], YY_train)
```

```python
# predict the test data
predictions = l1.predict(X[:, np.newaxis])
```

```python
# plot the real and predicted data
plt.axis([0, 1, -2, 2])
plt.plot(X, YY, color = 'green')
plt.plot(X,predictions)
plt.show() 
```
